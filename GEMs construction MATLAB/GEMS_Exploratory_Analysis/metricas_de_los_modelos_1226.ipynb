```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# =========================
# 1️⃣ Cargar los datos
# =========================
# Cambia el nombre del archivo si es CSV o Excel
df = pd.read_csv("Resumen_Modelos_ThermoKernel.csv")  # o pd.read_excel("tus_modelos.xlsx")

# Asegurarse de que las columnas numéricas son del tipo correcto
df['NumReacciones'] = pd.to_numeric(df['NumReacciones'])
df['SimilaridadPromedio'] = pd.to_numeric(df['SimilaridadPromedio'])

# =========================
# 2️⃣ Función para calcular todas las métricas
# =========================
def calcular_metricas(columna):
    metricas = {}
    metricas['Max'] = columna.max()
    metricas['Min'] = columna.min()
    metricas['Promedio'] = columna.mean()
    metricas['Mediana'] = columna.median()
    metricas['Moda'] = columna.mode().iloc[0] if not columna.mode().empty else np.nan
    metricas['Rango'] = columna.max() - columna.min()
    metricas['Varianza'] = columna.var()
    metricas['DesviacionEstandar'] = columna.std()
    metricas['CoeficienteVariacion'] = columna.std() / columna.mean() if columna.mean() != 0 else np.nan
    metricas['Q1'] = columna.quantile(0.25)
    metricas['Q3'] = columna.quantile(0.75)
    metricas['IQR'] = metricas['Q3'] - metricas['Q1']
    metricas['Percentil10'] = columna.quantile(0.10)
    metricas['Percentil90'] = columna.quantile(0.90)
    metricas['Asimetria'] = skew(columna)
    metricas['Curtosis'] = kurtosis(columna)
    return metricas

# =========================
# 3️⃣ Calcular métricas para NumReacciones y SimilaridadPromedio
# =========================
metricas_reacciones = calcular_metricas(df['NumReacciones'])
metricas_similitud = calcular_metricas(df['SimilaridadPromedio'])

# =========================
# 4️⃣ Proporciones de valores específicos
# =========================
metricas_reacciones['Proporcion_mas_1200'] = (df['NumReacciones'] > 1200).mean()
metricas_reacciones['Proporcion_menos_500'] = (df['NumReacciones'] < 500).mean()
metricas_similitud['Proporcion_mas_045'] = (df['SimilaridadPromedio'] > 0.45).mean()

# =========================
# 5️⃣ Mostrar resultados
# =========================
print("=== Métricas NumReacciones ===")
for k, v in metricas_reacciones.items():
    print(f"{k}: {v}")

print("\n=== Métricas SimilaridadPromedio ===")
for k, v in metricas_similitud.items():
    print(f"{k}: {v}")

# =========================
# 6️⃣ Opcional: guardar resultados en un Excel
# =========================
metricas_df = pd.DataFrame([metricas_reacciones, metricas_similitud], index=['NumReacciones', 'SimilaridadPromedio'])
metricas_df.to_excel("metricas_modelos.xlsx")
print("\nArchivo metricas_modelos.xlsx guardado.")
```
