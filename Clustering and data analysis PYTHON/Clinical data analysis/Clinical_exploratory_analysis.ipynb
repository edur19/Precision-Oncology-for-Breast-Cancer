# **Exploratory Analysis of Three Datasets**

The following three datasets correspond to the research patient cohort from the TCGA-BRCA project:

Clinical Data: Obtained from the GDC Data Portal (https://portal.gdc.cancer.gov/projects/TCGA-BRCA).

Survival Data: Obtained from the GDC Data Portal (https://portal.gdc.cancer.gov/projects/TCGA-BRCA).

Clinical Data (Portal Xena, California): Obtained from the Xena Browser.

Analysis 1

Clinical data obtained from:
https://portal.gdc.cancer.gov/projects/TCGA-BRCA

Dataset shape: (1255, 85)

Dataset variables by category:

1. Identification

sample → Sample ID of the analyzed specimen (e.g., TCGA-BH-A0W3-01A).

id → Unique UUID of the sample in GDC.

case_id → Unique identifier of the clinical case/patient.

submitter_id → ID assigned by the submitting center.

primary_site → Primary tumor site (e.g., Breast).

2. Patient Demographics

alcohol_history.exposures → Alcohol consumption history (Yes / No / Not Reported).

race.demographic → Race (White, Black, Asian, etc.).

gender.demographic → Gender (Female, Male).

ethnicity.demographic → Ethnicity (Hispanic/Latino, Not Hispanic or Latino).

vital_status.demographic → Vital status (Alive / Dead).

age_at_index.demographic → Age at the reference/index date.

days_to_birth.demographic → Days to birth (negative values).

year_of_birth.demographic → Year of birth.

year_of_death.demographic → Year of death (if applicable).

days_to_death.demographic → Days until death (if applicable).

3. TCGA Project Information

primary_site.project → Primary site as defined within the project.

project_id.project → Project ID (e.g., TCGA-BRCA).

disease_type.project → Associated neoplasm types.

name.project → Project name (e.g., Breast Invasive Carcinoma).

name.program.project → Program name (e.g., TCGA).

4. Sample Source Site

tissue_source_site_id.tissue_source_site → UUID of the source site.

code.tissue_source_site → Abbreviated site code (e.g., AR, BH).

name.tissue_source_site → Name of the submitting institution.

project.tissue_source_site → Project to which the sample belongs.

bcr_id.tissue_source_site → BCR file ID (Biospecimen Core Resource).

5. Annotations

entity_submitter_id.annotations → Submitter ID associated with the annotation (e.g., TCGA-A8-A08S).

notes.annotations → Clinical notes or comments (e.g., “No Note Specified”, “Patient had prior breast malignancy”).

submitter_id.annotations → Numeric submitter identifier (e.g., 815, 20712).

classification.annotations → Annotation category (e.g., Notification).

entity_id.annotations → Unique UUID for the annotation (e.g., 2779fa01-ac93-4e80-a997-3385f72172c3).

created_datetime.annotations → Creation timestamp (e.g., 2010-10-28T00:00:00).

annotation_id.annotations → Internal annotation ID (e.g., 2bf3041b-7f4f-5c9d-9dfb-f522ff0c6dae).

entity_type.annotations → Entity type linked to the annotation (e.g., case).

updated_datetime.annotations → Last update timestamp (e.g., 2018-08-23T16:36:27.780612-05:00).

case_id.annotations → UUID of the patient associated with the annotation.

state.annotations → System status (e.g., released).

category.annotations → Annotation category (e.g., Prior malignancy, Neoadjuvant therapy).

status.annotations → Approval status (e.g., Approved).

case_submitter_id.annotations → Patient ID used by the submitter (e.g., TCGA-A8-A08S).

6. Diagnosis

These fields describe clinical and pathological information about the patient and tumor.

Key diagnostic fields

synchronous_malignancy.diagnoses → Presence of synchronous malignancy (Yes/No)
Indicates whether another active cancer was present at the time of diagnosis.

ajcc_pathologic_stage.diagnoses → AJCC stage (Stage I–IV)
Based on TNM anatomical extent.

Stage I = small, localized tumors

Stage IV = metastatic disease

days_to_diagnosis.diagnoses → Days to diagnosis
Time from study entry or first symptom to confirmed diagnosis.

last_known_disease_status.diagnoses → Disease status
Examples: “not reported”, “progression”, “remission”.

tissue_or_organ_of_origin.diagnoses → Organ of origin
E.g., Breast, NOS.

days_to_last_follow_up.diagnoses → Days until last follow-up
Essential for survival analysis.

age_at_diagnosis.diagnoses → Age at diagnosis
Important prognostic factor.

primary_diagnosis.diagnoses → Main tumor diagnosis
E.g., “infiltrating duct carcinoma”.

prior_malignancy.diagnoses → Previous malignant neoplasms (Yes/No)

year_of_diagnosis.diagnoses → Diagnosis year

prior_treatment.diagnoses → Prior treatments (Yes/No)

ajcc_staging_system_edition.diagnoses → AJCC edition used

ajcc_pathologic_t.diagnoses → Tumor size (T1–T3)

morphology.diagnoses → Morphology code
Example: 8500/3 = invasive ductal carcinoma.

ajcc_pathologic_n.diagnoses → Lymph node status (N0–N2)

ajcc_pathologic_m.diagnoses → Metastasis (M0/M1)

classification_of_tumor.diagnoses → Clinical classification
E.g., carcinoma in situ, invasive, high-risk.

icd_10_code.diagnoses → ICD-10 code
Example: C50.9.

site_of_resection_or_biopsy.diagnoses → Biopsy/resection site

tumor_grade.diagnoses → Histologic grade (low/intermediate/high)

progression_or_recurrence.diagnoses → Tumor recurrence (Yes/No)

Derived fields (Xena)

age_at_earliest_diagnosis.diagnoses.xena_derived → Minimum diagnosis age

age_at_earliest_diagnosis_in_years.diagnoses.xena_derived → Earliest diagnosis age (years)

7. Treatments

treatment_id.treatments.diagnoses → Treatment ID(s).

submitter_id.treatments.diagnoses → Submitter treatment ID(s).

treatment_type.treatments.diagnoses → Treatment type (e.g., Pharmaceutical, Radiotherapy).

treatment_or_therapy.treatments.diagnoses → Treatment confirmation.

created_datetime.treatments.diagnoses → Creation date.

updated_datetime.treatments.diagnoses → Last update.

state.treatments.diagnoses → System status (e.g., released).

8. Biological Samples

sample_type_id.samples → Numeric sample type ID.

tumor_descriptor.samples → Tumor descriptor (Primary, Recurrent, etc.).

sample_id.samples → Sample UUID.

sample_type.samples → Sample type (Primary Tumor, Normal Tissue).

composition.samples → Tissue composition.

days_to_collection.samples → Days until collection.

initial_weight.samples → Initial weight.

preservation_method.samples → Preservation method (OCT, FFPE, etc.).

pathology_report_uuid.samples → Pathology report UUID.

oct_embedded.samples → OCT embedded (True/False).

specimen_type.samples → Specimen type (e.g., Solid Tissue).

days_to_sample_procurement.samples → Days until procurement.

is_ffpe.samples → Formalin-fixed/paraffin-embedded (True/False).

tissue_type.samples → Tumor / Normal.

annotations.samples → Sample annotations.

```r
# ============================================================
# DASHBOARD CLÍNICO TCGA-BRCA (VERSIÓN PYTHON COMPLETA)
# Conversión directa desde ggplot2 + patchwork (R)
# ============================================================

# =====================================
# 1. LIBRERÍAS
# =====================================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
import warnings

warnings.filterwarnings("ignore")

# =====================================
# 2. CONFIGURACIÓN GLOBAL
# =====================================

FONT_SIZE_INCREASE = 14
BASE_TITLE_FONT = 18 + FONT_SIZE_INCREASE
BASE_AXIS_TITLE_FONT = 12 + FONT_SIZE_INCREASE
BASE_AXIS_TICK_FONT = 10 + FONT_SIZE_INCREASE
BASE_SUPTITLE_FONT = 22 + FONT_SIZE_INCREASE

plt.rcParams.update({
    "axes.titlesize": BASE_TITLE_FONT,
    "axes.labelsize": BASE_AXIS_TITLE_FONT,
    "xtick.labelsize": BASE_AXIS_TICK_FONT,
    "ytick.labelsize": BASE_AXIS_TICK_FONT,
    "figure.facecolor": "white",
    "axes.facecolor": "white"
})

OUTPUT_DIR = Path("graficos_tcga_brca_pastel")
OUTPUT_DIR.mkdir(exist_ok=True)

PALETTES = ["cividis", "plasma", "magma", "viridis"]

VALUES_TO_UNIFY = {
    'not reported', 'unknown', 'nan', 'other',
    'no specific details', 'not applicable',
    '#n/a', 'n/a', 'na'
}

# =====================================
# 3. CARGA Y PREPROCESAMIENTO
# =====================================

def load_and_preprocess_data(file_path):
    df = pd.read_csv(file_path, sep="\t")
    print(f"✔️ Archivo cargado: {df.shape}")

    df_plot = df.copy()

    # Edad: días → años
    age_col = "age_at_diagnosis.diagnoses"
    if age_col in df_plot.columns:
        df_plot["age_at_diagnosis_corrected"] = (
            pd.to_numeric(df_plot[age_col], errors="coerce") / 365.25
        )
        print("✔️ Edad convertida a años")

    # Año diagnóstico
    year_col = "year_of_diagnosis.diagnoses"
    if year_col in df_plot.columns:
        df_plot[year_col] = pd.to_numeric(df_plot[year_col], errors="coerce")

    print(f"Dataset preparado: {df_plot.shape}")
    return df_plot


file_path = (
    "/Users/eduardoruiz/Documents/GitHub/"
    "Precision-Oncology-for-Breast-Cancer-Diagnosis/"
    "Clinical_data_and_models_ids/TCGA-BRCA.clinical.tsv"
)

df_plot = load_and_preprocess_data(file_path)

# =====================================
# 4. VARIABLES A GRAFICAR
# =====================================

VISUALIZATION_SPECS = [
    ("age_at_diagnosis_corrected", "Edad al Diagnóstico (Años)", "hist", None, "Edad (Años)"),
    ("gender.demographic", "Distribución de Género", "bar", None, None),
    ("race.demographic", "Distribución Racial", "bar", 10, None),
    ("vital_status.demographic", "Estado Vital", "bar", None, None),

    ("ajcc_pathologic_stage.diagnoses", "Estadios Patológicos AJCC", "bar", None, None),
    ("ajcc_pathologic_t.diagnoses", "Clasificación Tumor Primario (T)", "bar", None, None),
    ("ajcc_pathologic_n.diagnoses", "Clasificación Linfodos (N)", "bar", None, None),
    ("ajcc_pathologic_m.diagnoses", "Clasificación Metástasis (M)", "bar", None, None),

    ("progression_or_recurrence.diagnoses", "Progresión o Recurrencia", "bar", None, None),
    ("primary_diagnosis.diagnoses", "Diagnóstico Primario (Top 10)", "bar", 10, None),

    ("treatment_type.treatments.diagnoses", "Tipo de Tratamiento (Top 10)", "bar", 10, None),
    ("year_of_diagnosis.diagnoses", "Año de Diagnóstico", "hist", None, "Año"),
]

# =====================================
# 5. FUNCIONES AUXILIARES
# =====================================

def unify_missing(series):
    return (
        series.astype(str)
        .str.lower()
        .apply(lambda x: "Missing/Unknown"
               if x in VALUES_TO_UNIFY or x == "nan"
               else x)
    )


def create_plot(ax, df, spec, palette):
    col, title, plot_type, top_n, xlabel = spec

    if col not in df.columns:
        ax.text(0.5, 0.5, f"Columna '{col}' no encontrada",
                ha="center", va="center", color="gray")
        ax.axis("off")
        return

    if plot_type == "hist":
        data = df[col].dropna()
        if data.empty:
            ax.text(0.5, 0.5, "No hay datos válidos",
                    ha="center", va="center")
            ax.axis("off")
            return

        sns.histplot(
            data,
            bins=30,
            kde=True,
            color="#98AFC7",
            ax=ax
        )

    elif plot_type == "bar":
        counts = unify_missing(df[col]).value_counts()
        if top_n:
            counts = counts.head(top_n)

        sns.barplot(
            x=counts.index,
            y=counts.values,
            palette=palette,
            ax=ax
        )
        ax.set_ylabel("Conteo de Casos")
        ax.tick_params(axis='x', rotation=45)

    ax.set_title(title)
    ax.set_xlabel(xlabel if xlabel else col.replace("_", " ").title())

# =====================================
# 6. GENERACIÓN DE DASHBOARDS 2x3
# =====================================

plots_per_image = 6
num_images = int(np.ceil(len(VISUALIZATION_SPECS) / plots_per_image))

for img_idx in range(num_images):

    palette = PALETTES[img_idx % len(PALETTES)]

    fig, axes = plt.subplots(2, 3, figsize=(24, 16))
    axes = axes.flatten()

    start = img_idx * plots_per_image
    end = min(start + plots_per_image, len(VISUALIZATION_SPECS))

    for ax, spec in zip(axes, VISUALIZATION_SPECS[start:end]):
        create_plot(ax, df_plot, spec, palette)

    for ax in axes[end - start:]:
        ax.axis("off")

    fig.suptitle(
        f"Dashboard Clínico y Muestras Biológicas (Parte {img_idx + 1})",
        fontsize=BASE_SUPTITLE_FONT
    )

    plt.tight_layout(rect=[0, 0, 1, 0.95])

    filename = OUTPUT_DIR / f"dashboard_parte_{img_idx + 1}_{palette}.png"
    plt.savefig(filename, dpi=300)
    plt.show()

    print(f"✔️ Dashboard guardado: {filename}")

print(f"✅ Se generaron {num_images} dashboards correctamente.")
```

```r
# =====================================
# DASHBOARD TCGA-BRCA — Versión 4
# PDF final + letras más grandes
# =====================================

# ── 1. LIBRERÍAS ─────────────────────────────────────────────────────────────
required <- c("tidyverse", "ggplot2", "patchwork", "readr", "scales", "ggtext", "gridExtra", "grid")
invisible(lapply(required, library, character.only = TRUE))

# ── 2. CONFIGURACIÓN GLOBAL ──────────────────────────────────────────────────
OUTPUT_DIR <- "graficos_tcga_brca_v4"
if (!dir.exists(OUTPUT_DIR)) dir.create(OUTPUT_DIR)

PDF_OUT <- file.path(OUTPUT_DIR, "TCGA_BRCA_Dashboard_Completo.pdf")

PALETTES <- list(
  demografica   = "plasma",
  estadio       = "viridis",
  tratamiento   = "magma",
  muestra       = "cividis",
  supervivencia = "inferno"
)

MISSING_VALS <- c(
  'not reported', 'unknown', 'nan', 'other', 'no specific details',
  'not applicable', '#n/a', 'n/a', 'na', 'not available', "'--",
  'Not Reported', 'Unknown', 'Other', 'Not Available', 'not evaluated'
)

# ── TEMA BASE con letras grandes ─────────────────────────────────────────────
BASE_SIZE <- 18   # aumentado de 13 → 18

theme_dash <- theme_light(base_size = BASE_SIZE) +
  theme(
    plot.title       = element_text(size = BASE_SIZE + 5, face = "bold",
                                    margin = margin(b = 10), lineheight = 1.1),
    plot.subtitle    = element_text(size = BASE_SIZE + 1, color = "grey35",
                                    margin = margin(b = 8)),
    axis.title       = element_text(size = BASE_SIZE + 2, face = "bold"),
    axis.text        = element_text(size = BASE_SIZE),
    axis.text.x      = element_text(size = BASE_SIZE),
    axis.text.y      = element_text(size = BASE_SIZE, lineheight = 0.85),
    legend.text      = element_text(size = BASE_SIZE),
    legend.title     = element_text(size = BASE_SIZE + 1, face = "bold"),
    legend.position  = "none",
    panel.grid.minor = element_blank(),
    plot.margin      = margin(12, 65, 12, 12)
  )

# ── 3. CARGA Y PREPROCESAMIENTO ──────────────────────────────────────────────
`%||%` <- function(a, b) if (!is.null(a)) a else b

unify_missing <- function(x) {
  ifelse(is.na(x) | tolower(trimws(as.character(x))) %in% tolower(MISSING_VALS),
         "Missing/NA", as.character(x))
}

load_data <- function(file_path) {
  df <- readr::read_tsv(file_path, show_col_types = FALSE)

  if ("age_at_diagnosis.diagnoses" %in% colnames(df))
    df$age_years <- as.numeric(df$age_at_diagnosis.diagnoses) / 365.25

  if ("OS.time" %in% colnames(df))
    df$OS_time_years <- as.numeric(df$OS.time) / 365.25

  if ("year_of_diagnosis.diagnoses" %in% colnames(df))
    df$year_of_diagnosis.diagnoses <- as.numeric(df$year_of_diagnosis.diagnoses)

  if ("treatment_type.treatments.diagnoses" %in% colnames(df))
    df$treatment_type.treatments.diagnoses <- str_replace_all(
      df$treatment_type.treatments.diagnoses, "[\\[\\]']", "")

  if (all(c("oct_embedded.samples", "is_ffpe.samples") %in% colnames(df)))
    df$oct_ffpe_combined <- paste0(
      "OCT:", str_sub(as.character(df$oct_embedded.samples), 1, 1),
      " / FFPE:", str_sub(as.character(df$is_ffpe.samples), 1, 1))

  if ("OS" %in% colnames(df))
    df$OS_label <- case_when(
      is.na(df$OS)   ~ "Missing/NA",
      df$OS == 1     ~ "Fallecido (1)",
      df$OS == 0     ~ "Vivo (0)",
      TRUE           ~ "Missing/NA"
    )

  return(df)
}

file_path <- "/Users/eduardoruiz/Documents/GitHub/Precision-Oncology-for-Breast-Cancer-Diagnosis/Clinical_data_and_models_ids/TCGA-BRCA.clinical.tsv"
df <- load_data(file_path)
cat("✅ Dataset cargado:", nrow(df), "filas ×", ncol(df), "columnas\n")

# ── 4. FUNCIONES DE GRAFICACIÓN ──────────────────────────────────────────────

plot_hist <- function(df, col, title, subtitle = NULL, xlabel = col,
                      bins = 25, fill_col = "#4682B4") {
  df_c   <- df %>% filter(!is.na(!!sym(col)), is.finite(!!sym(col)))
  n_miss <- nrow(df) - nrow(df_c)

  p <- ggplot(df_c, aes(x = !!sym(col))) +
    geom_histogram(bins = bins, fill = fill_col, color = "white", alpha = 0.85) +
    stat_bin(bins = bins, geom = "text",
             aes(label = after_stat(ifelse(count > 0, count, ""))),
             vjust = -0.5, size = 5, fontface = "bold", color = "grey30") +
    scale_y_continuous(expand = expansion(mult = c(0, 0.20))) +
    labs(title    = title,
         subtitle = subtitle %||% paste0("n=", nrow(df_c),
                      if (n_miss > 0) paste0("  |  Missing: ", n_miss) else ""),
         x = xlabel, y = "Pacientes") +
    theme_dash

  if (str_detect(col, "year"))
    p <- p + scale_x_continuous(breaks = pretty_breaks(n = 8)) +
             theme(axis.text.x = element_text(angle = 45, hjust = 1, size = BASE_SIZE))
  p
}

plot_bar <- function(df, col, title, subtitle = NULL, top_n = Inf,
                     palette = "plasma", wrap_width = 28) {
  counts <- df %>%
    mutate(cat = unify_missing(!!sym(col))) %>%
    count(cat, sort = TRUE, name = "n")

  n_miss  <- sum(counts$n[counts$cat == "Missing/NA"], na.rm = TRUE)
  n_valid <- sum(counts$n[counts$cat != "Missing/NA"])

  if (top_n != Inf) {
    top_rows <- counts %>% filter(cat != "Missing/NA") %>% slice_head(n = top_n)
    miss_row <- counts %>% filter(cat == "Missing/NA")
    counts   <- bind_rows(top_rows, miss_row)
  }

  counts <- counts %>%
    mutate(
      pct      = round(n / nrow(df) * 100, 1),
      label    = paste0(n, " (", pct, "%)"),
      is_miss  = cat == "Missing/NA",
      cat_wrap = str_wrap(cat, width = wrap_width)
    ) %>%
    arrange(desc(is_miss), n) %>%
    mutate(cat_wrap = factor(cat_wrap, levels = unique(cat_wrap)))

  n_colors   <- sum(counts$cat != "Missing/NA")
  fill_vals  <- setNames(
    c(scales::viridis_pal(option = palette, begin = 0.25, end = 0.85)(max(n_colors, 1)),
      "grey75"),
    c(counts$cat_wrap[counts$cat != "Missing/NA"],
      counts$cat_wrap[counts$cat == "Missing/NA"])
  )

  ggplot(counts, aes(x = cat_wrap, y = n, fill = cat_wrap)) +
    geom_col(color = "white", width = 0.75) +
    geom_text(aes(label = label), hjust = -0.08,
              size = 5.5, fontface = "bold", color = "grey20") +
    coord_flip(clip = "off") +
    scale_fill_manual(values = fill_vals) +
    scale_y_continuous(expand = expansion(mult = c(0, 0.32))) +
    labs(title    = title,
         subtitle = subtitle %||% paste0("n válidos=", n_valid,
                      if (n_miss > 0) paste0("  |  Missing: ", n_miss) else ""),
         x = NULL, y = "Pacientes") +
    theme_dash
}

plot_binary_stacked <- function(df, col, title, subtitle = NULL,
                                labels_1 = "Sí (1)", labels_0 = "No (0)",
                                col1 = "#D55E00", col0 = "#0072B2") {
  df_p <- df %>%
    mutate(cat = case_when(
      is.na(!!sym(col))  ~ "Missing/NA",
      !!sym(col) == 1    ~ labels_1,
      !!sym(col) == 0    ~ labels_0,
      TRUE               ~ "Missing/NA"
    )) %>%
    count(cat) %>%
    mutate(pct   = round(n / nrow(df) * 100, 1),
           label = paste0(n, "\n(", pct, "%)"),
           cat   = factor(cat, levels = c(labels_0, labels_1, "Missing/NA")))

  fill_map <- setNames(c(col0, col1, "grey75"),
                       c(labels_0, labels_1, "Missing/NA"))

  ggplot(df_p, aes(x = "", y = n, fill = cat)) +
    geom_col(color = "white", width = 0.6) +
    geom_text(aes(label = label),
              position = position_stack(vjust = 0.5),
              size = 6, fontface = "bold", color = "white") +
    scale_fill_manual(values = fill_map, name = NULL) +
    scale_y_continuous(expand = expansion(mult = c(0, 0.05))) +
    coord_flip() +
    labs(title    = title,
         subtitle = subtitle %||% paste0("n=", nrow(df)),
         x = NULL, y = "Pacientes") +
    theme_dash +
    theme(legend.position = "right",
          legend.text     = element_text(size = BASE_SIZE),
          axis.text.y     = element_blank(),
          axis.ticks.y    = element_blank())
}

plot_survival_time <- function(df, col, title, xlabel = "Tiempo (años)") {
  df_p <- df %>%
    mutate(val      = as.numeric(!!sym(col)),
           has_data = !is.na(val) & is.finite(val))

  n_miss <- sum(!df_p$has_data)
  n_ok   <- sum(df_p$has_data)

  miss_df <- data.frame(
    Group = c("Con datos", "Missing/NA"),
    n     = c(n_ok, n_miss),
    pct   = round(c(n_ok, n_miss) / nrow(df) * 100, 1)
  ) %>% mutate(label = paste0(n, "\n(", pct, "%)"))

  p1 <- ggplot(df_p %>% filter(has_data), aes(x = val)) +
    geom_histogram(bins = 30, fill = "#4682B4", color = "white", alpha = 0.85) +
    stat_bin(bins = 30, geom = "text",
             aes(label = after_stat(ifelse(count > 0, count, ""))),
             vjust = -0.4, size = 4, color = "grey30") +
    scale_y_continuous(expand = expansion(mult = c(0, 0.20))) +
    labs(title    = title,
         subtitle = paste0("n con datos=", n_ok, "  |  Missing: ", n_miss),
         x = xlabel, y = "Pacientes") +
    theme_dash

  p2 <- ggplot(miss_df, aes(x = Group, y = n, fill = Group)) +
    geom_col(width = 0.55, color = "white") +
    geom_text(aes(label = label), vjust = -0.3,
              size = 5.5, fontface = "bold") +
    scale_fill_manual(values = c("Con datos" = "#4682B4", "Missing/NA" = "grey75")) +
    scale_y_continuous(expand = expansion(mult = c(0, 0.28))) +
    labs(title = "Completitud", x = NULL, y = "N") +
    theme_dash +
    theme(plot.title    = element_text(size = BASE_SIZE + 1, color = "grey40"),
          axis.text.x   = element_text(size = BASE_SIZE - 1))

  p1 + p2 + plot_layout(widths = c(3, 1))
}

# ── 5. ESPECIFICACIONES ───────────────────────────────────────────────────────
SPECS <- list(
  # DEMOGRAFÍA
  list("hist",   "age_years",                              "Edad al Diagnóstico",
       xlabel = "Años", fill_col = "#4682B4"),
  list("bar",    "gender.demographic",                     "Género",
       palette = PALETTES$demografica),
  list("bar",    "race.demographic",                       "Raza / Etnia (Top 10)",
       top_n = 10, palette = PALETTES$demografica),
  list("bar",    "ethnicity.demographic",                  "Etnicidad",
       palette = PALETTES$demografica),
  list("hist",   "year_of_diagnosis.diagnoses",            "Año de Diagnóstico",
       xlabel = "Año", fill_col = "#6A5ACD"),

  # ESTADIO Y TUMOR
  list("bar",    "ajcc_pathologic_stage.diagnoses",        "Estadio AJCC",
       palette = PALETTES$estadio),
  list("bar",    "ajcc_pathologic_t.diagnoses",            "Tumor Primario (T)",
       palette = PALETTES$estadio),
  list("bar",    "ajcc_pathologic_n.diagnoses",            "Linfonodos (N)",
       palette = PALETTES$estadio),
  list("bar",    "ajcc_pathologic_m.diagnoses",            "Metástasis (M)",
       palette = PALETTES$estadio),
  list("bar",    "tumor_grade.diagnoses",                  "Grado Tumoral",
       palette = PALETTES$estadio),
  list("bar",    "morphology.diagnoses",                   "Morfología (Top 10)",
       top_n = 10, palette = PALETTES$estadio),
  list("bar",    "primary_diagnosis.diagnoses",            "Diagnóstico Primario (Top 10)",
       top_n = 10, palette = PALETTES$estadio),

  # TRATAMIENTO
  list("bar",    "treatment_type.treatments.diagnoses",    "Tipo de Tratamiento (Top 10)",
       top_n = 10, palette = PALETTES$tratamiento),
  list("bar",    "treatment_or_therapy.treatments.diagnoses", "Terapia Recibida",
       palette = PALETTES$tratamiento),
  list("bar",    "prior_treatment.diagnoses",              "Tratamiento Previo",
       palette = PALETTES$tratamiento),
  list("bar",    "progression_or_recurrence.diagnoses",    "Progresión o Recurrencia",
       palette = PALETTES$tratamiento),
  list("binary", "Prior_Treatment_Flag",                   "Flag: Tratamiento Previo"),
  list("binary", "Metastasis_Flag",                        "Flag: Metástasis",
       labels_1 = "Con metástasis (1)", labels_0 = "Sin metástasis (0)"),

  # MUESTRA
  list("bar",    "sample_type.samples",                    "Tipo de Muestra",
       palette = PALETTES$muestra),
  list("bar",    "tissue_type.samples",                    "Tipo de Tejido",
       palette = PALETTES$muestra),
  list("bar",    "tumor_descriptor.samples",               "Descriptor Tumoral",
       palette = PALETTES$muestra),
  list("bar",    "specimen_type.samples",                  "Tipo de Espécimen",
       palette = PALETTES$muestra),
  list("bar",    "oct_ffpe_combined",                      "Conservación: OCT / FFPE",
       palette = PALETTES$muestra),
  list("binary", "is_ffpe.samples",                        "Muestra FFPE",
       labels_1 = "FFPE (TRUE)", labels_0 = "No FFPE (FALSE)"),
  list("binary", "oct_embedded.samples",                   "Embedded en OCT",
       labels_1 = "OCT (TRUE)", labels_0 = "No OCT (FALSE)"),
  list("bar",    "site_of_resection_or_biopsy.diagnoses",  "Sitio de Biopsia (Top 10)",
       top_n = 10, palette = PALETTES$muestra),

  # SUPERVIVENCIA
  list("binary", "OS_label",                               "Estado de Supervivencia (OS)",
       labels_1 = "Fallecido (1)", labels_0 = "Vivo (0)"),
  list("surv",   "OS_time_years",                          "Tiempo de Supervivencia",
       xlabel = "Años desde diagnóstico"),
  list("bar",    "vital_status.demographic",               "Estado Vital",
       palette = PALETTES$supervivencia),
  list("bar",    "last_known_disease_status.diagnoses",    "Último Estado de Enfermedad",
       palette = PALETTES$supervivencia),
  list("bar",    "alcohol_history.exposures",              "Historial de Alcohol",
       palette = PALETTES$supervivencia)
)

# ── 6. BUILD INDIVIDUAL PLOT ─────────────────────────────────────────────────
build_plot <- function(df, spec) {
  tipo  <- spec[[1]]
  col   <- spec[[2]]
  title <- spec[[3]]
  args  <- spec[-(1:3)]

  if (!col %in% colnames(df)) {
    cat("⚠️  Columna no encontrada:", col, "\n")
    return(ggplot() +
             annotate("text", x = 0.5, y = 0.5,
                      label = paste("Sin datos:\n", col),
                      size = 7, color = "grey60") +
             theme_void() +
             labs(title = title) +
             theme(plot.title = element_text(size = BASE_SIZE + 3, face = "bold",
                                             color = "grey50")))
  }

  tryCatch({
    if (tipo == "hist") {
      plot_hist(df, col, title,
                xlabel   = args$xlabel   %||% col,
                bins     = args$bins     %||% 25,
                fill_col = args$fill_col %||% "#4682B4")

    } else if (tipo == "bar") {
      plot_bar(df, col, title,
               top_n   = args$top_n   %||% Inf,
               palette = args$palette %||% "plasma")

    } else if (tipo == "binary") {
      plot_binary_stacked(df, col, title,
                          labels_1 = args$labels_1 %||% "Sí (1)",
                          labels_0 = args$labels_0 %||% "No (0)",
                          col1     = args$col1     %||% "#D55E00",
                          col0     = args$col0     %||% "#0072B2")

    } else if (tipo == "surv") {
      plot_survival_time(df, col, title,
                         xlabel = args$xlabel %||% "Tiempo")
    }
  }, error = function(e) {
    cat("❌ Error en columna", col, ":", conditionMessage(e), "\n")
    ggplot() +
      annotate("text", x = 0.5, y = 0.5,
               label = paste("Error:\n", col), size = 7, color = "red") +
      theme_void() + labs(title = title)
  })
}

# ── 7. SECCIONES ─────────────────────────────────────────────────────────────
secciones <- list(
  list(name = "Sección 1 — Características Demográficas",  idx = 1:5),
  list(name = "Sección 2 — Estadio y Biología Tumoral",    idx = 6:12),
  list(name = "Sección 3 — Historia Terapéutica",          idx = 13:18),
  list(name = "Sección 4 — Características de la Muestra", idx = 19:26),
  list(name = "Sección 5 — Supervivencia y Estado Clínico",idx = 27:31)
)

# ── 8. GENERAR PDF COMPLETO ───────────────────────────────────────────────────
cat("\n📄 Generando PDF completo:", PDF_OUT, "\n")

pdf(PDF_OUT, width = 22, height = 15, onefile = TRUE)

# ── PORTADA ──────────────────────────────────────────────────────────────────
grid.newpage()
grid.rect(gp = gpar(fill = "#1a1a2e", col = NA))
grid.text("TCGA-BRCA",
          x = 0.5, y = 0.78,
          gp = gpar(fontsize = 52, fontface = "bold", col = "white"))
grid.text("Dashboard de Análisis Clínico Completo",
          x = 0.5, y = 0.65,
          gp = gpar(fontsize = 32, col = "#a0c4ff"))
grid.text(paste0("Total de muestras: ", nrow(df),
                 "  |  Variables graficadas: ", length(SPECS)),
          x = 0.5, y = 0.50,
          gp = gpar(fontsize = 24, col = "grey70"))
grid.text(paste0("Secciones: Demografía · Estadio · Tratamiento · Muestra · Supervivencia"),
          x = 0.5, y = 0.40,
          gp = gpar(fontsize = 20, col = "grey60"))
grid.text(paste0("Generado: ", format(Sys.time(), "%d %b %Y")),
          x = 0.5, y = 0.28,
          gp = gpar(fontsize = 18, col = "grey50"))

# ── PÁGINAS POR SECCIÓN ──────────────────────────────────────────────────────
plots_per_page <- 4

for (sec in secciones) {
  sec_specs  <- SPECS[sec$idx]
  sec_title  <- sec$name

  # Página de separador de sección
  grid.newpage()
  grid.rect(gp = gpar(fill = "#f0f4f8", col = NA))
  grid.text(sec_title,
            x = 0.5, y = 0.55,
            gp = gpar(fontsize = 40, fontface = "bold", col = "#1a1a2e"))
  grid.text(paste0(length(sec_specs), " variables  |  TCGA-BRCA  |  n=", nrow(df)),
            x = 0.5, y = 0.42,
            gp = gpar(fontsize = 24, col = "grey50"))

  # Construir plots
  all_plots <- map(sec_specs, ~build_plot(df, .x)) %>%
    Filter(Negate(is.null), .)

  if (length(all_plots) == 0) next

  n_pages <- ceiling(length(all_plots) / plots_per_page)

  for (pg in seq_len(n_pages)) {
    start      <- (pg - 1) * plots_per_page + 1
    end        <- min(start + plots_per_page - 1, length(all_plots))
    page_plots <- all_plots[start:end]

    while (length(page_plots) < plots_per_page)
      page_plots[[length(page_plots) + 1]] <- ggplot() + theme_void()

    combined <- wrap_plots(page_plots, ncol = 2, nrow = 2) +
      plot_annotation(
        title    = sec_title,
        subtitle = paste0("Parte ", pg, " de ", n_pages,
                          "  |  TCGA-BRCA  |  n=", nrow(df), " muestras"),
        theme    = theme(
          plot.title    = element_text(size = BASE_SIZE + 14, hjust = 0.5,
                                       face = "bold", margin = margin(b = 6)),
          plot.subtitle = element_text(size = BASE_SIZE + 4, hjust = 0.5,
                                       color = "grey40")
        )
      )

    print(combined)
    cat("  📊 Sección:", sec_title, "| Página", pg, "de", n_pages, "\n")
  }
}

dev.off()
cat("\n✅ PDF generado:", PDF_OUT, "\n")
cat("   Total páginas aprox.:", 1 + length(secciones) +
    sum(sapply(secciones, function(s) ceiling(length(s$idx) / plots_per_page))), "\n")
```

```r
library(ggplot2)
library(dplyr)
library(gridExtra)
library(RColorBrewer)
library(fs)
library(readxl)

# -------------------------------
# 1️⃣ Leer el archivo
archivo <- "/Users/eduardoruiz/Documents/GitHub/Precision-Oncology-for-Breast-Cancer-Diagnosis/Clinical_data_and_models_ids/MetaData.xlsx" 
df <- read_excel(archivo)
colnames(df) <- gsub(" ", "_", colnames(df))

# -------------------------------
# 2️⃣ Columnas categóricas
columnas_categoricas <- c("Race", "Definition", "Menopausal_Status", "Cancer_Type", 
                          "ER", "PR", "HER2", "Event", "Event_Site", 
                          "Subtype", "Genetic_Ancestry")
df_cat <- df %>% select(any_of(columnas_categoricas))

# -------------------------------
# 3️⃣ Crear carpeta para guardar dashboards
carpeta_dashboard <- "dashboard_graficas_estetico"
dir_create(carpeta_dashboard)

# -------------------------------
# 4️⃣ Función para gráficos 2x2
graficos_2x2 <- function(cols, df){
  lista_graficos <- lapply(cols, function(cn){
    valores <- df[[cn]]
    n <- length(unique(na.omit(valores)))
    colores <- RColorBrewer::brewer.pal(min(max(n,3), 12), "Set3")
    colores <- rep(colores, length.out=n)
    
    ggplot(df, aes(x = .data[[cn]], fill = .data[[cn]])) +
      geom_bar() +
      geom_text(stat="count", aes(label=..count..), vjust=-0.1, size=6) +
      scale_fill_manual(values = colores) +
      labs(title = cn, x=NULL, y="Número de pacientes") +
      theme_minimal(base_size = 16) +
      theme(axis.text.x = element_text(angle=45, hjust=1, size=18),
            axis.text.y = element_text(size=18),
            axis.title.y = element_text(size=20),
            plot.title = element_text(face="bold", size=20),
            legend.position = "none")
  })
  
  archivo_salida <- file.path(carpeta_dashboard, paste0("dashboard_", paste(cols, collapse="_"), ".png"))
  png(archivo_salida, width=16, height=12, units="in", res=300)
  do.call(grid.arrange, c(lista_graficos, ncol=2, nrow=2))
  dev.off()
  print(paste0("✅ Dashboard guardado: ", archivo_salida))
}

# -------------------------------
# 5️⃣ Función para gráficos individuales (muchas categorías) horizontal
grafico_individual <- function(col_name, df){
  valores <- df[[col_name]]
  n <- length(unique(na.omit(valores)))
  colores <- RColorBrewer::brewer.pal(min(max(n,3), 12), "Set3")
  colores <- rep(colores, length.out=n)
  
  p <- ggplot(df, aes(x = .data[[col_name]], fill = .data[[col_name]])) +
    geom_bar() +
    geom_text(stat="count", aes(label=..count..), vjust=-0.5, size=6) +
    scale_fill_manual(values = colores) +
    labs(title = col_name, x=NULL, y="Número de pacientes") +
    coord_flip() +  # Horizontal
    theme_minimal(base_size = 16) +
    theme(axis.text.x = element_text(size=18),
          axis.text.y = element_text(size=18),
          axis.title.y = element_text(size=20),
          plot.title = element_text(face="bold", size=20),
          legend.position = "none")
  
  archivo_salida <- file.path(carpeta_dashboard, paste0("grafico_individual_", col_name, ".png"))
  ggsave(archivo_salida, plot = p, width=14, height=10, dpi=300)
  print(paste0("✅ Gráfico individual guardado: ", archivo_salida))
}

# -------------------------------
# 6️⃣ Generar dashboards 2x2
grupos_2x2 <- list(
  c("Race", "Definition", "Menopausal_Status", "ER"),
  c("PR", "HER2", "Subtype", "Genetic_Ancestry"),
  c("Event") # Último grupo pequeño
)

for (grupo in grupos_2x2){
  graficos_2x2(grupo, df_cat)
}

# -------------------------------
# 7️⃣ Gráficos individuales para muchas categorías (horizontal)
grafico_individual("Cancer_Type", df_cat)
grafico_individual("Event_Site", df_cat)
```

Search of clinical match of the metabolic models ids to its clinical data

```r
import pandas as pd
import numpy as np
import re
import os

# =======================================================
# 0. CONFIGURACIÓN
# =======================================================
OUTPUT_DIR = "analisis_clasificacion_modelos"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Definición de la columna de ID de muestra completo en df_clinical
COL_SAMPLE_ID = 'sample'
COL_SAMPLE_TYPE = 'sample_type.samples'

# =======================================================
# 1. Carga de Datos Clínicos (Base de 1255 filas)
# =======================================================
df_clinical = pd.read_csv("TCGA-BRCA.clinical.tsv", sep="\t")
print("Base clínica total cargada:", len(df_clinical))


# =======================================================
# 2. CARGA DE LA LISTA DE MODELOS REALES 🔑
# =======================================================

try:
    # Leemos el archivo y extraemos la primera columna
    df_model_names = pd.read_csv('lista_modelos_unicos.txt', header=None, sep='\t', skipinitialspace=True)

    # Convertimos la columna a una lista de strings
    lista_modelos_unicos = df_model_names.iloc[:, 0].dropna().astype(str).tolist()

except FileNotFoundError:
    print("\n❌ ERROR: Archivo 'lista_modelos_unicos.txt' no encontrado. No se puede continuar.")
    lista_modelos_unicos = [] # Detener la ejecución si no se encuentra el archivo
    exit()

print(f"Archivos de modelos únicos cargados: {len(lista_modelos_unicos)}")


# =======================================================
# 3. Extracción de IDs y Creación del DataFrame de Modelos
# =======================================================

# Función para limpiar el nombre del archivo y obtener el ID de muestra (ej: TCGA-3C-AAAU-01A)
def extract_sample_id(filename):
    # Busca el patrón TCGA-XX-XXXX-XXA y lo extrae antes de cualquier guion bajo
    match = re.search(r'(TCGA-[A-Z0-9]{2}-[A-Z0-9]{4}-[A-Z0-9]{2}[A-Z0-9])', filename)
    if match:
        return match.group(0)
    return filename.split('_')[0].strip()

# Aplicar la función a la lista de nombres de archivos
model_sample_ids = [extract_sample_id(model) for model in lista_modelos_unicos]

# Crear un DataFrame con los IDs de modelo
df_modelos = pd.DataFrame({
    COL_SAMPLE_ID: model_sample_ids,
    'modelo_path_completo': lista_modelos_unicos, # Mantener el nombre completo
})

# ❌ LÍNEA ELIMINADA: df_modelos.drop_duplicates(...)
# Ahora, si lista_modelos_unicos tiene 1226 elementos, df_modelos tendrá 1226 elementos.

# Mostrar el conteo de IDs ÚNICOS para verificar la repetición
print(f"IDs de muestra ÚNICOS extraídos de los modelos: {df_modelos[COL_SAMPLE_ID].nunique()}")
print(f"Filas totales del set de Modelos (Base, NO FILTRADO): {len(df_modelos)}")

# =======================================================
# 4. Clasificación de los IDs de Modelos (Merge de Clasificación)
# =======================================================

# 4.1. Crear la tabla de referencia (lookup) con ID de Muestra y Tipo de Muestra
df_lookup_sample_type = df_clinical[[COL_SAMPLE_ID, COL_SAMPLE_TYPE]].copy()

# 4.2. Merge del set de Modelos con la tabla de Clasificación Clínica
# Este merge unirá la clasificación de la muestra a cada línea de modelo.
df_modelos_clasificados = pd.merge(
    df_modelos,
    df_lookup_sample_type,
    on=COL_SAMPLE_ID,
    how='left' # Usar left-merge para mantener todos los modelos
)

# =======================================================
# 5. Análisis y Resumen del Set de Modelos Clasificados
# =======================================================

print("\n" + "="*70)
print("## 📊 Clasificación de Muestras Basada en el Set de Modelos")
print("="*70)

print("### 📝 Conteo de Tipos de Muestra para el Set de Modelos:")
classification_counts = df_modelos_clasificados[COL_SAMPLE_TYPE].value_counts(dropna=False)
print(classification_counts)

# Verificación
samples_not_found = df_modelos_clasificados[df_modelos_clasificados[COL_SAMPLE_TYPE].isna()]

print("\n### ⚠️ Muestras de Modelos No Encontradas en Datos Clínicos:")
if len(samples_not_found) > 0:
    print(f"Se encontraron {len(samples_not_found)} IDs de modelos que no tienen match en los datos clínicos de TCGA.")
    print("Primeros 5 IDs no encontrados:")
    print(samples_not_found[COL_SAMPLE_ID].head())
else:
    print("✅ Todos los IDs de modelos se encontraron en la base de datos clínica.")

# =======================================================
# 6. Definición del Set Final de 1226 Filas
# =======================================================
df_final_modelos = df_modelos_clasificados.copy()

print("\n--- PASO FINAL ---")
print(f"El DataFrame 'df_final_modelos' ahora contiene la clasificación de sus {len(df_final_modelos)} modelos.")
```
